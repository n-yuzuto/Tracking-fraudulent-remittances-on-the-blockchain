```python
import requests
import json
import pprint
```

```python
def main():
    URL = 'https://rinkeby.infura.io/v3/9a530eac50ea405c9c3985d96259ca06'
    H = "0xa43da49e5aa8caaec07de19c0884a21ac3dba567fabaf34da406ce0c3b92aff3"
    headers = {'content-type': 'application/json'}
    data = {"jsonrpc":"2.0","method":"eth_getTransactionByHash","params": [H],"id":4}
    
    
    result = requests.post(URL,headers=headers,data=json.dumps(data))
    pprint.pprint(result.json())

if __name__ == '__main__':
    main()
```

```python
def main():
    #初期設定
    URL = 'https://rinkeby.infura.io/v3/9a530eac50ea405c9c3985d96259ca06'
    HEADERS = {'content-type': 'application/json'}
    number = 4103834
    address_list = ["0xd0e0a5c78fcc0f9d6a0a509e13645bf86b1e365f"]

    
    #heightを増やして犯人のアドレスを見る
    for i in range(102):
        data = {"jsonrpc":"2.0","method":"eth_getBlockByNumber","params": [hex(number),True],"id":4}
        r = requests.post(URL,headers=HEADERS,data=json.dumps(data))
        d = r.json()["result"]["transactions"]
        if i == 0:
            print("Transactions:")
        for j in d:
            if j["from"] in address_list:
                address_list.append(j["to"])
                if i == 0:
                    continue
                value=int(j["value"],0) / 1000000000000000000
                print("from: {}, to: {}".format(j["from"],j["to"]))
                print("send: {} eth".format(value))
        
        number = number + 1
    print("Number of Account: {}".format(int(len(address_list))-1))
    print("Accounts:")
    pprint.pprint(address_list)
    
if __name__ == '__main__':
    main()
```
